# ANÁLISIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('3_HR_DATA_nulos.csv')
```

```python
df.columns
```

### 1. Análisis Univariante

```python
# Variables numéricas principales
numeric = [
    'monthly_income_$', 'age', 'total_working_years',
    'years_at_company', 'distance_from_home',
    'percent_salary_hike' 
]

# Variables categóricas clave
categorical = ['attrition', 'over_time', 'remote_work', 'worklife_balance', 'relationship_satisfaction', 'marital_status','gender']

# Estadísticas básicas para numéricas
display(df[numeric].describe().T)
display(df[categorical].describe(include = "object").T)

```

```python
# Histogramas y boxplots
for i in numeric:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1) # Parámetros: (filas, columnas, índice de subgráfico)
    sns.histplot(df[i], kde=True, bins=30)
    plt.title(f'Distribución de {i}')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[i])
    plt.title(f'Boxplot de {i}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Gráficos de barras para categóricas
for i in categorical:
    if i in df.columns:  # Verifica que la variable exista en el DataFrame
        plt.figure(figsize=(6, 4))
        # Cuenta los valores, elimina nulos, calcula proporciones
        prop = df[i].value_counts(normalize=True) 
        prop.plot(kind='bar')
        plt.title(f'Frecuencia relativa de {i}')
        plt.ylabel('Proporción')
        plt.xticks(rotation=45)
        plt.grid(axis='y', linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()
    else:
        print(f'Variable {i} no encontrada en el DataFrame.')
```

```python
# Gráfico de quesito para visualizar mejor el porcentaje de la variable attrition
attrition_counts = df['attrition'].value_counts()
plt.pie(attrition_counts, labels= attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=['#5DADE2', '#F7DC6F'])  # autopct muestra los porcentajes
plt.title('Porcentaje de Attrition')
plt.show()
```

```python
#OPCIÓN ANILLO PARA LA PRESENTACIÓN:

attrition_counts = df['attrition'].value_counts()
colors = ['#4F81BD', '#FCD116']  # Azul y amarillo

# Crear gráfico de donut
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    attrition_counts, 
    labels=attrition_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.4),  # Esto crea el donut
    textprops=dict(color="black", fontsize=12)
)

# Estilo del porcentaje (autotexto)
for autotext in autotexts:
    autotext.set_fontweight('bold')
    autotext.set_color('black')

# Título
ax.set_title('Porcentaje de Attrition', fontsize=14, fontweight='bold')

plt.show()
```

### 2. Análisis Bivariante

#### Todas las categorías vs Attrition

```python
# Histogramas y boxplots para las variables numéricas agrupadas por 'attrition'
for i in numeric:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Boxplot
    sns.boxplot(data=df, x='attrition', y=i, palette='Set2', ax=axes[0])
    axes[0].set_title(f'{i} vs. Attrition')
    axes[0].grid(axis='y', linestyle='--', alpha=0.5)
    # Histograma (agrupado por attrition para más información)
    sns.histplot(data=df, x=i, hue='attrition', kde=False, palette='Set2', element='step', ax=axes[1])
    axes[1].set_title(f'Histogram of {i} for Attrition')
    axes[1].grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# Barras apiladas para categóricas vs. attrition
for i in categorical:
    crosstab = pd.crosstab(df[i], df['attrition'], normalize='index')
    crosstab.plot(kind='bar', stacked=False, figsize=(7, 5), colormap='Set3') # Si eligiéramos stacked=True, las barras se apilan
    plt.title(f'Attrition por {i}')
    plt.ylabel('Proporción')
    plt.legend(title='Attrition', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

#### Percent salary hike vs years at company

```python
# Histogramas y boxplots para las variables numéricas agrupadas por 'attrition'
fig, axes = plt.subplots(1, 2, figsize=(20, 5))
    
# Boxplot
sns.boxplot(data=df, x='years_at_company', y='percent_salary_hike', palette='Set2', ax=axes[0])
axes[0].set_title('Percent salary hike vs years at company')
axes[0].grid(axis='y', linestyle='--', alpha=0.5)

# Histograma (agrupado por attrition para más información)
sns.histplot(data=df, x='percent_salary_hike', hue='attrition', kde=False, palette='Set2', element='step', ax=axes[1])
axes[1].set_title(f'Histogram of percent salary hike for Attrition')
axes[1].grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Generamos una matriz de correlación

correlation_matrix = df.select_dtypes(include='number').corr()
correlation_matrix
```

#### Perfiles de interés

Perfiles de riesgo potencialmente alto y constatación o no según tasa de rotación actual
Perfiles a cuidar, etc

```python
# SIN INTERÉS

# Jóvenes con baja satisfacción

not_important = df[
    (df['age'] < df['age'].quantile(0.25)) &
    (df['environment_satisfaction'] <= 2)
]

print(f"Número de empleados de este perfil: {len(not_important)}")
print("\nTasa de rotación en jóvenes con baja satisfacción:\n", not_important['attrition'].value_counts(normalize=True).round(2))

# No se constata que sea un perfil de riesgo alto (sólo 32% renuncian)
```

```python
# ¡¡POZO NEGRO DE FUGAS!!

# Personas jóvenes, que menos cobran, que llevan poco tiempo en la empresa y que hacen horas extra
young_escaping = df[
    (df['monthly_income_$'] < df['monthly_income_$'].quantile(0.25)) &
    (df['years_at_company'] < 2) &
    (df['over_time'] == 'Yes') &
    (df['age'] < df['age'].quantile(0.25))
]

print(f"Número de empleados de este perfil: {len(young_escaping)}")
print("Tasa de rotación en personas menores de", int(df['age'].quantile(0.25)), ", nuevas en la empresa, con los salarios más bajos y que hacen horas extra:\n", young_escaping['attrition'].value_counts(normalize=True).round(2))

# Se confirma que el perfil tiene un riesgo altísimo (93% renuncian)
```

```python
# CHOLLO PARA LA EMPRESA. Ninguna persona de este grupo se está yendo sorprendentemente ¿se podría deducir la importancia del job satisfaction?, en cualquier caso, sin job satisfaction (siguiente celda) se van un poco más pero el 89% se queda.

# Personas con un puesto de responsabilidad alta, satisfechas con sus funciones y que menos cobran salarios bajos
happy_gold = df[
    (df['monthly_income_$'] < df['monthly_income_$'].quantile(0.50)) &
    (df['job_satisfaction'] > 2) &
    (df['job_level'] > df['job_level'].quantile(0.50))
]

print(f"Número de empleados de este perfil: {len(happy_gold)}")
print(f"Tasa de rotación en personas con un puesto de responsabilidad alta, llegada reciente a la empresa, satisfechas con sus funciones y que menos cobran salarios bajos:\n", happy_gold['attrition'].value_counts(normalize=True).round(2))
```

```python
# CHOLLO PARA LA EMPRESA. Cuidar que no se vayan ¿aumentando su job satisfaction por ejemplo?

# Personas con un puesto de responsabilidad alta, y que menos cobran salarios bajos
gold = df[
    (df['monthly_income_$'] < df['monthly_income_$'].quantile(0.50)) &
    (df['job_level'] > df['job_level'].quantile(0.50))
]

print(f"Número de empleados de este perfil: {len(gold)}")
print(f"Tasa de rotación en personas con un puesto de responsabilidad alta, llegada reciente a la empresa, satisfechas con sus funciones y que menos cobran salarios bajos:\n {gold['attrition'].value_counts(normalize=True).round(2)}, \n{gold['job_satisfaction'].value_counts(normalize=True).round(2)}")
```

```python
# Personas veteranas con puestos de responsabilidad de nivel medio-bajo y baja satisfacción

# Umbrales:
antiguedad_media = df[df['job_level'] == 2]['years_at_company'].median()
satisfaccion_baja = df['job_satisfaction'].quantile(0.25)

# Filtrar el grupo
obsoletes = df[
    (df['job_level'] == 2) &
    (df['years_at_company'] >= antiguedad_media) &
    (df['job_satisfaction'] <= satisfaccion_baja)
]


print(f"Número de empleados de este perfil: {len(obsoletes)}")
print(f"Tasa de rotación en personas veteranas con puestos de responsabilidad de nivel medio-bajo y baja satisfacción:\n {(obsoletes['attrition'].value_counts(normalize=True).round(2))}")
```

```python
# Personas de responsabilidad media-alta, antigüedad media ó alta, alta satisfacción con sus funciones: 

# Umbrales
antiguedad_media = df[df['job_level'] == 3]['years_at_company'].median()
satisfaccion_alta = df['job_satisfaction'].quantile(0.75)

# Filtrar el grupo
intermediate_satisfied = df[
    (df['job_level'] == 3) &
    (df['years_at_company'] >= antiguedad_media) &
    (df['job_satisfaction'] >= satisfaccion_alta)
]


print(f"Número de empleados de este perfil: {len(intermediate_satisfied)}")
print(f"Tasa de rotación en personas de responsabilidad media-alta, antigüedad >= {antiguedad_media}, satisfacción >= {satisfaccion_alta}:\n {intermediate_satisfied['attrition'].value_counts(normalize=True).round(2)}")
```

### Dispersión entre variables

```python
# Gráfico de: salario, años en la empresa, grado de responsabilidad y satisfacción con las funciones

plt.figure(figsize=(15, 7))
sns.scatterplot(
    data=df,
    x='monthly_income_$',
    y='years_at_company',
    marker= 'o',
    hue= 'job_level',
    size='job_satisfaction',
    palette='Set1',
    alpha=0.7
)
plt.title('Salario, años en la empresa, grado de responsabilidad y satisfacción con las funciones')

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de: salario, años en la empresa, grado de responsabilidad y satisfacción con las funciones de las personas que han renunciado
# (es el mismo gráfico que el anterior filtrando sólo las personas que ya han renunciado)

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=df[df['attrition'] =='yes'],
    x='monthly_income_$',
    y='years_at_company',
    marker= 'o',
    hue= 'job_level',
    size='job_satisfaction',
    palette='Set1',
    alpha=0.7
)
plt.title('Salario, años en la empresa, grado de responsabilidad y satisfacción con las funciones de las personas que han renunciado')

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de: salario, años en la empresa, grado de responsabilidad y satisfacción con las funciones de las personas que siguen en la empresa

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=df[df['attrition'] =='no'],
    x='monthly_income_$',
    y='years_at_company',
    marker= 'o',
    hue= 'job_level',
    size='job_satisfaction',
    palette='Set1',
    alpha=0.7
)
plt.title('Salario, años en la empresa, grado de responsabilidad y satisfacción con las funciones de las personas que siguen en la empresa')

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,7))

# Filtrar por categoría
yes_df = df[df['attrition'] == 'yes']
no_df = df[df['attrition'] == 'no']

# Scatterplot para "yes"
sns.scatterplot(
    data=yes_df,
    x='monthly_income_$',
    y='age',
    marker='o',
    color='red',
    label='Attrition: Yes',
    alpha=0.7
)

# Scatterplot para "no"
sns.scatterplot(
    data=no_df,
    x='monthly_income_$',
    y='age',
    marker='x',
    color='blue',
    label='Attrition: No',
    alpha=0.7
)

plt.title('Salario vs. Edad, diferenciado por Attrition')
plt.xlabel('Monthly Income ($)')
plt.ylabel('Age')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
